(1)簡易每日記帳總額計算
程式需計算並輸出：
1. 總支出金額
2. 總共記錄了幾筆支出
(2)
範例1：
輸入：
100
50
200
0

輸出：
總支出：350 元
支出筆數：3 筆

範例2：
輸入：
30
70
0

輸出：
總支出：100 元
支出筆數：2 筆
(3)本程式可以幫助使用者快速記錄每日花費總額，
讓使用者清楚知道一天花了多少錢。

透過簡單的輸入方式，可以培養記帳習慣。
(4)本程式使用了以下Python語法與概念：

1. while 迴圈：持續輸入資料直到輸入0為止
2. if 判斷式：判斷是否結束輸入
3. input()：接收使用者輸入
4. float() 或 int()：轉換數值型態
5. 變數累加：計算總支出與筆數
6. print()：輸出結果
(5)本題為自行設計，將日常生活中的記帳行為簡化成程式操作，
讓使用者透過輸入支出金額來統計每日花費。


```python
total = 0   # 用來累計總支出
count = 0   # 用來計算筆數

print("請輸入每筆支出金額（輸入0 結束）：")

while True:
    expense = float(input("輸入金額："))

    # 如果輸入0，代表結束
    if expense == 0:
        break

    # 累加總支出
    total += expense

    # 筆數加1
    count += 1

# 輸出結果
print("總支出：", int(total), "元")
print("支出筆數：", count, "筆")
```
